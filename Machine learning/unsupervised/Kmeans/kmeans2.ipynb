```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
sns.set()
```

```python
data=pd.read_csv("Countries-exercise.csv")
data
```

```python
x=data.iloc[:,1:3]
x
```

```python
kmeans=KMeans(32)
kmeans.fit(x)
```

```python
data1=data.copy()
data1["cluster"]=kmeans.fit_predict(x)
data1
```

```python
plt.scatter(data1["Longitude"],data1["Latitude"],c=data1["cluster"],cmap="multi")
plt.xlim(-250,250)
plt.ylim(-120,120)
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```

```python
WCSS=[]
for i in range(1,241):
    kmean=KMeans(i)
    kmean.fit(x)
    wcss_iter=kmean.inertia_
    WCSS.append(wcss_iter)
    
```

```python
WCSS
```

```python
cluster_no=range(1,241)
plt.plot(cluster_no,WCSS)
plt.xlabel("no of clusters")
plt.ylabel("WCSS")
plt.show()
```
