```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
raw_data=pd.read_csv("bank_data.csv")
raw_data
```

```python
raw_data=raw_data.drop(["Unnamed: 0"],axis=1)
raw_data
```

```python
raw_data["y"]=raw_data["y"].map({"yes":1,"no":0})
raw_data
```

```python
y=raw_data["y"]
x1=raw_data.drop(["y"],axis=1)
```

```python
x=sm.add_constant(x1)
reg_logg=sm.Logit(y,x)
result=reg_logg.fit()
```

```python
result.summary()
```

```python
np.set_printoptions(formatter={'float':lambda x:"{0:0.2f}".format(x)})
dataframe=pd.DataFrame(data=result.predict(x),columns=["predicted"])
```

```python
dataframe["y"]=raw_data["y"]
dataframe
```

```python
dataframe["predicted"]=dataframe["predicted"].round()
```

```python
def comfusion_matrix(data,actual_values,model):
    pred_value=model.predict(data)
    bins=np.array([0,0.5,1])
    cm=np.histogram2d(actual_values,pred_value,bins=bins)[0]
    accurate=(cm[0,0]+cm[1,1])/cm.sum()
    return cm,accurate
```
