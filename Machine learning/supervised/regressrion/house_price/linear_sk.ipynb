```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.linear_model import LinearRegression
```

```python
data=pd.read_csv("real_estate_sklearn.csv")
```

```python
data.head()
```

```python
x=data["size"]
y=data["price"]
```

```python
x.shape
```

```python
y.shape
```

```python
x_matrix=x.values.reshape(-1,1)
x_matrix
```

```python
x_matrix.shape
```

```python
lin_reg=LinearRegression()
```

```python
lin_reg.fit(x_matrix,y)
```

```python
lin_reg.score(x_matrix,y)
```

```python
lin_reg.coef_
```

```python
lin_reg.intercept_
```

```python
new_data=pd.DataFrame(data=[1234],columns=["size"])
matrix=new_data["size"].values.reshape(-1,1)
matrix
```

```python
new_data["prediction"]=lin_reg.predict(matrix)
new_data
```

```python
plt.scatter(x,y)
line=lin_reg.intercept_+(lin_reg.coef_*x)
fig=plt.plot(x,line,lw=4,c="red")
plt.xlabel("size")
plt.ylabel("prize")
plt.show()
```
