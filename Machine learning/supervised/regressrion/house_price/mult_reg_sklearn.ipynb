### FORMULA FOR ADJUSTED R SQUARE
$R^2_{adj.}=1-(1-R^2)*\frac{n-1}{n-p-1}$

```python
r2=reg.score(x,y)
n=x.shape[0]
p=x.shape[1]
```

```python
adj_r=1-(1-r2)*(n-1)/(n-p-1)
adj_r
```
