```python
data_with_dummy=pd.get_dummies(no_milcolinear_data,drop_first=True)
data_with_dummy
```

```python
data_with_dummy.columns.values
```

```python
cols=['Log_price', 'Mileage', 'EngineV', 'Brand_BMW',
       'Brand_Mercedes-Benz', 'Brand_Mitsubishi', 'Brand_Renault',
       'Brand_Toyota', 'Brand_Volkswagen', 'Body_hatch', 'Body_other',
       'Body_sedan', 'Body_vagon', 'Body_van', 'Engine Type_Gas',
       'Engine Type_Other', 'Engine Type_Petrol', 'Registration_yes']


    
```

```python
data_preprocessed=data_with_dummy[cols]
data_preprocessed=data_preprocessed.astype(float)
data_preprocessed.describe()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
variable=data_preprocessed[['Mileage', 'EngineV', 'Brand_BMW',
       'Brand_Mercedes-Benz', 'Brand_Mitsubishi', 'Brand_Renault',
       'Brand_Toyota', 'Brand_Volkswagen', 'Body_hatch', 'Body_other',
       'Body_sedan', 'Body_vagon', 'Body_van', 'Engine Type_Gas',
       'Engine Type_Other', 'Engine Type_Petrol', 'Registration_yes']]
vif2=pd.DataFrame()
vif2["VIF"]=[variance_inflation_factor(variable,i) for i in range(variable.shape[1])]
vif2["Features"]=variable.columns
```

```python
vif2
```

```python
target=data_preprocessed["Log_price"]
Features=data_preprocessed.drop(["Log_price"],axis=1)
```

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
scaler.fit(Features)
```

```python
scaled_data=scaler.transform(Features)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(scaled_data,target,test_size=0.2,random_state=365)
y_train
```

```python
reg=LinearRegression()
reg.fit(x_train,y_train)
```

```python
y_hat=reg.predict(x_train)
```

```python
plt.scatter(y_train,y_hat)
plt.xlabel("targers(y_train)")
plt.ylabel("Predicted(y_hat)")
plt.xlim(6,13)
plt.ylim(6,13)
```

```python
##RESIDUALS PLOTe
```

```python
sns.distplot(y_train-y_hat)
```

```python
reg.score(x_train,y_train)
```
